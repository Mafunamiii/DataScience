```python
#Load the required packages and read in the training data set as adult_tr
#1 FILL IN THE BLANK
import pandas as pd

#2 FILL IN THE BLANK
import numpy as np
import statsmodels.tools.tools as stattools

#3 FILL IN THE BLANK
from sklearn.tree import DecisionTreeClassifier, export_graphviz

#4 ADDRESS OF DATASET
adult_tr = pd.read_csv("/home/mafunami/Windows_SSD_250/Users/Jahn/PythonSRC/Week_7/adult_ch6_training")
```

```python
#5 FILL IN THE BLANK - For simplicity, we save the Income variable. THE PREVIOUS SENTENCE ALREADY GIVEN THE VARIABLE NAME.
y = adult_tr[['Income']]
```

```python
#6 We have a categorical (variable type) variable, Marital status, among our predictors. 
#7 The CART model implemented in the sklearn package needs variables converted to a dummy variable form. 
#8 We save the matrix and dictionary separately using (mar_cat, mar_cat_dict).

mar_np = np.array(adult_tr['Marital status'])
(mar_cat, mar_cat_dict) = stattools.categorical(mar_np, drop=True, dictnames = True)
```

```python
#The matrix mar_cat contains five columns, one for each category in the original Marital status variable. 
#Each row represents a record in the adult_tr data set. 
#Each row will have a 1 in the column which matches the value that record had
#in the original Marital status variable. You can tell which column represents which
#category by examining mar_cat_dict. 

#9 FILL IN THE BLANK
mar_cat_pd = pd.DataFrame(mar_cat)
```

```python
#10 FILL THE BLANK - the command is synonymous to the word conjugate
X = pd.concat((adult_tr[['Cap_Gains_Losses']], mar_cat_pd), axis = 1)

#We first make the mar_cat matrix 
#We attach the predictor variableCap_Gains_Losses to the data frame of dummy variables that represent marital status.
X_names = ["Cap_Gains_Losses", "Divorced", "Married", "Never-married", "Separated", "Widowed"]
```

```python
#11 FILL IN THE BLANK-It will help us when visualizing the CART model to know the levels of y as well

y_names = ["<=50K", ">50K"]
```

```python
#12 Now, we are ready to run the algorithm! FILL IN THE BLANK

cart01 = DecisionTreeClassifier(criterion = "gini", max_leaf_nodes=5) .fit(X,y)
data = export_graphviz(cart01, out_file=None, feature_names=X_names, class_names=y_names)
```

```python
#17 Next, load the required libraries.
from sklearn.ensemble import RandomForestClassifier
import numpy as np
```

```python
#Python requires a response variable formatted as a one‐dimensional array. What numpy command is used to create that format?

#18 FILL IN THE BLANK
rfy = np.ravel(y)

#19 FILL IN THE BLANK
rf01 = RandomForestClassifier(n_estimators = 100, criterion="gini") .fit(X, rfy)
rf01.predict(X)
```
